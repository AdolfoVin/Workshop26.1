##1. Atribuição das bibliotecas que serão utilizadas

```python
import sys
import scipy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from pandas.plotting import scatter_matrix
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

##2. declaração da tabela de dados que serão extraídos e polidos




```python
df = pd.read_csv('/content/Video Games Sales (1980-2024) - Raw.csv')
df.head(20)
```

```python
df = df.drop(['img','title', 'publisher', 'developer', 'release_date', 'last_update'], axis=1, errors='ignore')
df['is_hit'] = df['total_sales'].fillna(0).astype(int)
```

```python
#Removendo as linahs com valores vazios (garante menos instabilidade de dados)
df = df.dropna()
```

```python
#Nesse sessão entramos na parte de polimento "avançado", servirá como pré processamento dos dados, transformando os dados relevantes todos na forma de números
le = LabelEncoder()
df['genre'] = le.fit_transform(df['genre'])
df['console'] = le.fit_transform(df['console'])
```

```python
#Separação dos dados (x representa nossos valores que serão usados como 'variáveis' enquanto o y será nosso target)
x = df.drop(['is_hit', 'total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales'], axis=1)
y = df['is_hit']
```

```python
#Divisão treino/teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Padronização
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

##3. Fazendo o treinamento do algoritmo


```python
modelo = RandomForestClassifier(n_estimators=100, random_state=42)
modelo.fit(x_train, y_train)
```

##4. Testando o algoritmo


```python
previsoes = modelo.predict(x_test)
acuracia = accuracy_score(y_test, previsoes)

print(f"Acurácia do modelo: {acuracia * 100:.2f}%")
print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, previsoes))
print("\nRelatório de Classificação:")
print(classification_report(y_test, previsoes))
```

```python

#Salvando a solução temos:
joblib.dump(modelo, 'modelo_de_vendas_dos_videogames.pkl')
print("\nModelo salvo com ducesso!")
```
