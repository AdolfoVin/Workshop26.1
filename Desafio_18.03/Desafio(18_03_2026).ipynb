###1-Adição das bibliotecas e cração das variáveis de temperatura e dias da semana

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import random as np

Temp = np.randint(20, 35, size=(1, 7))
days = ['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo']
```

###2- Criação do DataFrame usando o pandas

```python
df = pd.DataFrame(Temp, columns = days)
df.head()
```

###3- Média das temperaturas

```python
media = float(sum(Temp[0])/7)
print('A média das temperaturas é de ', media, 'ºC')
```

###4- Temperaturas máxima e mínima

```python
for max in Temp:
  print('A temperatura máxima é de ', max.max(), 'ºC')
for min in Temp:
  print('A temperatura mínima é de ', min.min(), 'ºC')
```

###5- Plotando gráficos baseados nas variáveis criadas

```python
eixox = days
eixoy = Temp[0]

plt.plot(eixox, eixoy, label="Temperatura semanal")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura em ºC")
plt.title("Temperatura na semana")
plt.style.use('fast')
plt.xticks(rotation=45)
plt.show()
```

```python
eixox = days
eixoy = Temp[0]

plt.stem(eixox, eixoy, label="Temperatura semanal")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura em ºC")
plt.title("Temperatura na semana")
plt.style.use('petroff10')
plt.xticks(rotation=45)
plt.show()
```

```python
eixox = days
eixoy = Temp[0]

plt.bar(eixox, eixoy, label="Temperatura semanal")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura em ºC")
plt.title("Temperatura na semana")
plt.style.use('petroff10')
plt.xticks(rotation=45)
plt.show()
```
